```python
%%js

# Factors of 50: 1, 2, 5, 10, 25, 50

for num in range(1, 51):
    print(f"\nNumber {num}:")
    
    # First nested conditional - check divisibility by 1
    if num % 1 == 0:
        print(f"  Divisible by 1")
        
        # Second nested conditional - check divisibility by 2
        if num % 2 == 0:
            print(f"  Divisible by 2")
            
            # Third nested conditional - check divisibility by 5
            if num % 5 == 0:
                print(f"  Divisible by 5")
                
                # Fourth nested conditional - check divisibility by 10
                if num % 10 == 0:
                    print(f"  Divisible by 10")
                    
                    # Fifth nested conditional - check divisibility by 25
                    if num % 25 == 0:
                        print(f"  Divisible by 25")
                        
                        # Sixth nested conditional - check divisibility by 50
                        if num % 50 == 0:
                            print(f"  Divisible by 50")
                            print(f"  *** {num} is divisible by ALL factors of 50! ***")
                        else:
                            print(f"  Not divisible by 50")
                    else:
                        print(f"  Not divisible by 25")
                else:
                    print(f"  Not divisible by 10")
            else:
                print(f"  Not divisible by 5")
        else:
            print(f"  Not divisible by 2")
```
